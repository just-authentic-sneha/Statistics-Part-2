# Statistics Part 2

### 1. What is hypothesis testing in statistics?
Hypothesis testing is a statistical method used to make decisions or inferences about population parameters based on sample data. It involves formulating a null hypothesis and an alternative hypothesis, then using sample data to determine whether to reject the null hypothesis.

---

### 2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
The null hypothesis (H₀) is a statement that there is no effect or no difference, and it is tested for possible rejection. The alternative hypothesis (H₁ or Ha) represents a new claim or effect that we want to test. Rejecting H₀ supports the alternative hypothesis.

---

### 3. What is the significance level in hypothesis testing, and why is it important?
The significance level (denoted by α) is the threshold for rejecting the null hypothesis. It represents the probability of making a Type I error (false positive). Common values are 0.05 or 0.01. It helps control the rate of false positives in hypothesis testing.

---

### 4. What does a P-value represent in hypothesis testing?
A P-value is the probability of observing the sample data, or something more extreme, under the assumption that the null hypothesis is true. It quantifies the evidence against the null hypothesis.

---

### 5. How do you interpret the P-value in hypothesis testing?
- If the P-value ≤ α: Reject the null hypothesis (evidence supports H₁).
- If the P-value > α: Fail to reject the null hypothesis (insufficient evidence to support H₁).

---

### 6. What are Type 1 and Type 2 errors in hypothesis testing?
- **Type I error**: Rejecting the null hypothesis when it is actually true (false positive).
- **Type II error**: Failing to reject the null hypothesis when it is actually false (false negative).

---

### 7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
- **One-tailed test**: Tests for a deviation in one specific direction (greater or less).
- **Two-tailed test**: Tests for a deviation in either direction from the hypothesized value.

---

### 8. What is the Z-test, and when is it used in hypothesis testing?
A Z-test is used to determine if there is a significant difference between sample and population means when the population standard deviation is known and the sample size is large (n > 30).

---

### 9. How do you calculate the Z-score, and what does it represent in hypothesis testing?
**Z-score formula**:  
\[ Z = \frac{\bar{x} - \mu}{\sigma / \sqrt{n}} \]  
Where:
- \( \bar{x} \) = sample mean  
- \( \mu \) = population mean  
- \( \sigma \) = population standard deviation  
- \( n \) = sample size  

It measures how many standard deviations the sample mean is from the population mean.

---

### 10. What is the T-distribution, and when should it be used instead of the normal distribution?
The T-distribution is used instead of the normal distribution when the sample size is small (n < 30) and the population standard deviation is unknown. It has heavier tails to account for increased uncertainty.

---

### 11. What is the difference between a Z-test and a T-test?
- **Z-test**: Used when population standard deviation is known and sample size is large.
- **T-test**: Used when population standard deviation is unknown and/or sample size is small.

---

### 12. What is the T-test, and how is it used in hypothesis testing?
A T-test compares the means of one or two groups and determines whether the difference is statistically significant. It is used when the population standard deviation is unknown.

---

### 13. What is the relationship between Z-test and T-test in hypothesis testing?
Both tests assess whether the sample data provide sufficient evidence to reject the null hypothesis. A T-test is a generalization of the Z-test, used when more uncertainty exists (i.e., unknown standard deviation or smaller sample size).

---

### 14. What is a confidence interval, and how is it used to interpret statistical results?
A confidence interval provides a range of values within which the true population parameter is expected to lie, with a given level of confidence (e.g., 95%). It helps quantify uncertainty in parameter estimates.

---

### 15. What is the margin of error, and how does it affect the confidence interval?
The margin of error is the amount added and subtracted from the sample estimate to create the confidence interval. A larger margin of error results in a wider interval, reflecting greater uncertainty.

---

### 16. How is Bayes' Theorem used in statistics, and what is its significance?
Bayes' Theorem allows updating the probability of a hypothesis based on new evidence. It is used in inferential statistics, machine learning, and decision-making to revise beliefs as more data becomes available.

---

### 17. What is the Chi-square distribution, and when is it used?
The Chi-square distribution is a skewed distribution used in tests of independence and goodness-of-fit. It is commonly used when dealing with categorical data.

---

### 18. What is the Chi-square goodness of fit test, and how is it applied?
The Chi-square goodness of fit test determines whether observed frequencies differ significantly from expected frequencies. It is used to test if a sample matches a population distribution.

---

### 19. What is the F-distribution, and when is it used in hypothesis testing?
The F-distribution is used to compare two variances and in ANOVA tests. It is right-skewed and depends on two degrees of freedom (numerator and denominator).

---

### 20. What is an ANOVA test, and what are its assumptions?
ANOVA (Analysis of Variance) tests whether there are statistically significant differences between the means of three or more groups. Assumptions include:
- Independence of observations
- Normality of the data
- Equal variances across groups (homoscedasticity)

---

### 21. What are the different types of ANOVA tests?
- **One-way ANOVA**: Tests one independent variable across multiple groups.
- **Two-way ANOVA**: Tests two independent variables and their interaction.
- **Repeated measures ANOVA**: Used when the same subjects are measured multiple times.

---

### 22. What is the F-test, and how does it relate to hypothesis testing?
The F-test compares the variances of two populations or tests the overall significance in ANOVA. It evaluates whether group means differ more than expected by chance, using the F-distribution.

```python
#1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.
import scipy.stats as stats
import numpy as np

sample = [2.1, 2.5, 2.8, 3.0, 2.6]
population_mean = 2.5
population_std = 0.2
n = len(sample)
sample_mean = np.mean(sample)
z = (sample_mean - population_mean) / (population_std / np.sqrt(n))
p_value = 2 * (1 - stats.norm.cdf(abs(z)))

print(f"Z-score: {z}")
print(f"P-value: {p_value}")

if p_value < 0.05:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

```python
#2. Simulate random data to perform hypothesis testing and calculate the corresponding P-value using Python.
np.random.seed(0)
sample_data = np.random.normal(loc=100, scale=15, size=50)
population_mean = 95
z_score = (np.mean(sample_data) - population_mean) / (np.std(sample_data, ddof=1) / np.sqrt(len(sample_data)))
p_val = 2 * (1 - stats.norm.cdf(abs(z_score)))

print(f"Z-score: {z_score}")
print(f"P-value: {p_val}")
```

```python
#3. Implement a one-sample Z-test using Python to compare the sample mean with the population mean.
def one_sample_z_test(sample, pop_mean, pop_std):
    sample_mean = np.mean(sample)
    z = (sample_mean - pop_mean) / (pop_std / np.sqrt(len(sample)))
    p = 2 * (1 - stats.norm.cdf(abs(z)))
    return z, p

data = [105, 110, 98, 115, 102]
z_val, p_val = one_sample_z_test(data, 100, 10)
print(f"Z-score: {z_val}, P-value: {p_val}")
```

```python
#5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.
def visualize_errors(mu0=100, mu1=105, sigma=15, n=30, alpha=0.05):
    se = sigma / np.sqrt(n)
    critical = stats.norm.ppf(1 - alpha)
    beta = stats.norm.cdf((critical * se + mu0 - mu1) / se)
    power = 1 - beta

    x = np.linspace(mu0 - 3*se, mu1 + 3*se, 1000)
    y0 = stats.norm.pdf(x, mu0, se)
    y1 = stats.norm.pdf(x, mu1, se)

    plt.plot(x, y0, label='H0 Distribution')
    plt.plot(x, y1, label='H1 Distribution')
    plt.axvline(mu0 + critical * se, color='red', linestyle='--', label='Critical Value')

    plt.fill_between(x, y0, where=x > mu0 + critical * se, color='red', alpha=0.3, label='Type I Error')
    plt.fill_between(x, y1, where=x <= mu0 + critical * se, color='blue', alpha=0.3, label='Type II Error')

    plt.legend()
    plt.title('Type I and Type II Errors')
    plt.xlabel('Sample Mean')
    plt.ylabel('Density')
    plt.show()

visualize_errors()
```

```python
#9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    stderr = stats.sem(data)
    margin = stats.t.ppf((1 + confidence) / 2., n-1) * stderr
    return mean - margin, mean + margin

sample_data = np.random.normal(100, 15, 50)
ci = confidence_interval(sample_data)
print(f"95% Confidence Interval: {ci}")
```

```python
#10. Write a Python program to calculate the margin of error for a given confidence level using sample data.
def margin_of_error(data, confidence=0.95):
    n = len(data)
    stderr = stats.sem(data)
    t_score = stats.t.ppf((1 + confidence) / 2, n-1)
    return t_score * stderr

sample_data = np.random.normal(100, 15, 30)
moe = margin_of_error(sample_data)
print(f"Margin of Error: {moe}")
```
